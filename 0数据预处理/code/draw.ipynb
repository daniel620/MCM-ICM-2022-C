```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

d = pd.read_excel('/Users/kuzaowuwei/Desktop/2022美赛/赛中数据/0数据预处理/data/price_cleaned.xlsx')
g = d['usdpm'].values
b = d['value'].values
date = d['date'].values
def plot_gold(is_show=True):
    plt.plot(range(len(g)),g)
    plt.xticks([i for i in range(len(date))], date,rotation=45)
    plt.ylabel('value of portfolio')
    plt.xlabel('date')
    plt.legend()
    x_major_locator=plt.MultipleLocator(180)
    ax=plt.gca()
    ax.xaxis.set_major_locator(x_major_locator)
    plt.xlim(0)
    plt.ylim(0,2500)
    ax.ticklabel_format(style='sci', scilimits=(-1,2), axis='y')
    # plt.title('{}'.format(portfolio_value[-1]))
    if is_show==True:
        plt.show()    
    return 

def plot_bitcoin(is_show=True):
    plt.plot(range(len(b)),b)
    plt.xticks([i for i in range(len(date))], date,rotation=45)
    plt.ylabel('value of portfolio')
    plt.xlabel('date')
    plt.legend()
    x_major_locator=plt.MultipleLocator(180)
    ax=plt.gca()
    ax.xaxis.set_major_locator(x_major_locator)
    plt.xlim(0)
    plt.ylim(0,70000)
    ax.ticklabel_format(style='sci', scilimits=(-1,2), axis='y')
    # plt.title('{}'.format(portfolio_value[-1]))
    if is_show==True:
        plt.show()
    return
```

```python
plot_gold(False)
plot_bitcoin()
```
