```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel('/Users/kuzaowuwei/Desktop/2022美赛/赛中数据/1预测模型/data/AR3阶价格预测.xlsx')
value = data['value'][3:]
pred_value = data['pred_value'][3:]
```

```python
n = len(value)
t = np.arange(n)
plt.subplot(2,1,1)
plt.plot(t, value)
plt.ylabel("value")

plt.subplot(2,1,2)
plt.plot(t, pred_value, 'r')
plt.ylabel('pred_value')
plt.tight_layout()
plt.show()
```

```python
n = len(value)
t = np.arange(n)
plt.plot(t,value,'b',linewidth=0.2)
plt.ylabel("value")

plt.plot(t,pred_value,'r',linewidth=0.2)
```
